## Working with pivot tables

```python
# importing pandas
import pandas as pd

# importing sales dataset
temperatures = pd.read_csv("../datasets/temperatures.csv")
temperatures.head()
```

#### Pivoting the dog pack

```python
temp_avg_temp_by_country_vs_date = temperatures.pivot_table(values="avg_temp_c", index="country", columns="date")
temp_avg_temp_by_country_vs_date
```

#### .loc[] + slicing is a power combo

```python
temp_avg_temp_by_country_vs_date.loc["China":"Taiwan"]
```

#### The axis argument

```python
temp_avg_temp_by_country_vs_date.mean(axis="index")
```

#### Calculating summary stats across columns

```python
temp_avg_temp_by_country_vs_date.mean(axis="columns")
```
