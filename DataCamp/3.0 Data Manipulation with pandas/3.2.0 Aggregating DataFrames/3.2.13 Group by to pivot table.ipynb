## Aggregating DataFrames

#### Pivoting on one variable

#### Avoiding

```python
# importing pandas
import pandas as pd

# importing sales dataset
temp = pd.read_csv("../datasets/temperatures.csv")
temp.head()
```

#### Group by to pivot table

```python
temp.groupby("country")["avg_temp_c"].mean()
```

```python
temp.pivot_table(values="avg_temp_c", index="country")
```

#### Different statistics

```python
import numpy as np
temp.pivot_table(values="avg_temp_c", index="country", aggfunc=np.median)
```

```python
temp.pivot_table(values="avg_temp_c", index="country", aggfunc=[np.mean, np.median])
```

#### Pivot on two variables

```python
temp.groupby(["country", "city"])["avg_temp_c"].mean()
```

```python
temp.pivot_table(index="country", columns="city", values="avg_temp_c")
```

```python
temp.pivot_table(values="avg_temp_c", index="country", columns="city", fill_value=0)
```

#### Summing with pivot tables

```python
temp.pivot_table(values="avg_temp_c", index="country", columns="city", fill_value=0, margins=True)
```
