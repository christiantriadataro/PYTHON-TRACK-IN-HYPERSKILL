## Visualizing DataFrames

#### Changes in sales over time

Line plots are designed to visualize the relationship between two numeric variables, where each data values is connected to the next one. They are especially useful for visualizing the change in a number over time since each time point is naturally connected to the next time point. In this exercise, you'll visualize the change in avocado sale over three years. `pandas` has been imported as `pd`, and `avocados` is available.

```python
# importing pandas
import pandas as pd

# importing sales dataset
avocados = pd.read_pickle("../datasets/avoplotto.pkl")
avocados.head()
```
