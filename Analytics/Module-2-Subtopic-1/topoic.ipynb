### Cleaning Empty Cells

#### B. If you want to change the original DataFrame, use the inplace=True argument to remove all rows with NULL values:

```python
import pandas as pd

df = pd.read_csv("data.csv")
df.dropna(inplace=True)
print(df.to_string())
```

#### C. Replace Empty Values

```python
import pandas as pd

df = pd.read_csv("data.csv")
df.fillna(130, inplace=True)
print(df.to_string())
```

#### D. Replace only be specified columns

```python
import pandas as pd

pd = pd.read_csv("data.csv")
df["Calories"].fillna(130, inplace=True)
print(df.to_string())
```

#### E. Replace using mean, median or mode
#### Calculate the mean and replace any empty values with it

```python
import pandas as pd
x = df["Calories"].mean()
df["Calories"].fillna(x, inplace=True)
print("Mean:", x)
print(df.to_string())
```

#### Calculate the Median, and replace any empty values with it:

```python
import pandas as pd
x = df["Calories"].median()
df["Calories"].fillna(x, inplace=True)
print("Median:", x)
print(df.to_string())
```

#### Calculate the Mode, and replace any empty values with it:

```python
import pandas as pd
x = df["Calories"].mode()[0]
df["Calories"].fillna(x, inplace=True)
print("Mode:", x)
print(df.to_string())
```

#### 3. Cleaning the wrong data

```python
import pandas as pd

df = pd.read_csv("data.csv")
df.loc[7, "Duration"] = 45
print(df.to_string())
```

Loop through all values in the "Duration" column:

```python
for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.loc[x, "Duration"] = 120
print(df.to_string())
```
